```python
!pip install textblob ipywidgets

from textblob import TextBlob
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

def determine_sentiment(query):
  """Определение тональности запроса."""
  analysis = TextBlob(query)
  if analysis.sentiment.polarity >= 0:
      return "positive"
  return "negative"

def respond_as_batman():
  """Ответы в стиле Бэтмена."""
  batman_quotes = [
        "It's not who I am underneath, but what I do that defines me.",
        "I am vengeance. I am the night. I am Batman!",
        "Sometimes the truth isn't good enough, sometimes people deserve more."
    ]
  return random.choice(batman_quotes)

def respond_as_joker():
  """Ответы в стиле Джокера."""
  joker_quotes = [
        "Why so serious?",
        "We stop looking for monsters under the bed when we realize they are inside us.",
        "I believe whatever doesn’t kill you, simply makes you... stranger."
    ]
  return random.choice(joker_quotes)

def process_query(query):
  """Главная функция, обрабатывающая запросы."""
  sentiment = determine_sentiment(query)
  if sentiment == "positive":
    return respond_as_batman()
  return respond_as_joker()

# Примеры запросов для тестирования.
test_queries = [
    "I love the way you think!",
    "This is the worst day ever.",
    "I can't wait for the weekend!",
    "Why is everything so difficult?",
    "You are my hero.",
    "I'm so tired of this nonsense.",
    "What a beautiful day!",
    "Nothing ever goes right for me.",
    "I am so excited about this!",
    "I can't stand the traffic."
]

def on_button_click(b):
  """Обработка пользовательского ввода."""
  with output:
      clear_output()
      query = text_input.value
      response = process_query(query)
      print(f"Query: {query}\nResponse: {response}\n")

# Создание интерактивных элементов
text_input = widgets.Text(
    value='',
    placeholder='Введите ваш запрос...',
    description='Запрос:',
    disabled=False
)

button = widgets.Button(
    description='Отправить',
    disabled=False,
    button_style='',
    tooltip='Нажмите для отправки запроса',
    icon='check'
)

output = widgets.Output()

# Связываем нажатие кнопки с функцией обработки
button.on_click(on_button_click)

# Отображаем элементы
display(text_input, button, output)

# Проведение тестов.
results = [(query, process_query(query)) for query in test_queries]

# Отображение результатов.
for query, response in results:
    print(f"Query: {query}\nResponse: {response}\n")
```
